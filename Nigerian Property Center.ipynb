```python
property_df = {
    'property':[],
    'address':[],
    'price':[],
    'duration':[]
}
```

```python
for i in range(1,120):
    url = f'https://nigeriapropertycentre.com/for-rent?page={i}'
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    page = soup.find(class_ = "property-list-page")
    apartments =page.find_all(class_ = 'wp-block-content clearfix text-xs-left text-sm-left')
    
    

    for apartment in apartments:
          property = apartment.find('a').text
          address = apartment.find('address', class_ = 'voffset-bottom-10').text
          #description = apartment.find(class_ = 'description hidden-xs').text
          price = apartment.find('span', class_ = 'price').find_next('span').text
          duration = apartment.find(class_ = 'period').text


          property_df['property'].append(property)
          property_df['address'].append(address)
          property_df['price'].append(price)
          property_df['duration'].append(duration) 
    
      
  
```

```python
amenities_dict = {
    'number_of_bedrooms':[],
    'number_of_bathrooms':[]
}  
```

```python
for i in range(1,120):
    url = f'https://nigeriapropertycentre.com/for-rent?page={i}'
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    page = soup.find(class_ = "property-list-page")
    amenities = page.find_all(class_ = 'wp-block-footer')
    
    
    for amenity in amenities:
        number_of_bedrooms_element = amenity.find('span')
        number_of_bathrooms_element = number_of_bedrooms_element.find_next('span') if number_of_bedrooms_element else None

        number_of_bedrooms = number_of_bedrooms_element.text if number_of_bedrooms_element else 'unknown'
        number_of_bathrooms = number_of_bathrooms_element.find_next('span').text if number_of_bathrooms_element else 'unknown'



        amenities_dict['number_of_bedrooms'].append(number_of_bedrooms)
        amenities_dict['number_of_bathrooms'].append(number_of_bathrooms)     
```

```python
amenities_df = pd.DataFrame(amenities_dict)
```
